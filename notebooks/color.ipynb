```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import imutils
```

```python
image1 = cv2.imread('/sky.png')
```

```python
plt.imshow(image1)
```

```python
imageRGB = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imageRGB)
```

```python
color = ('r','g','b')
for i,col in enumerate(color):
    histr = cv2.calcHist([imageRGB],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
imageHSV = cv2.cvtColor(imageRGB, cv2.COLOR_RGB2HSV)
```

```python
plt.imshow(imageHSV)
```

```python
hue = imageHSV[:,:,0]
sat = imageHSV[:,:,1]
val = imageHSV[:,:,2]
plt.figure(figsize=(8,12))
plt.subplot(311)
plt.imshow(hue,cmap=plt.cm.gray)
plt.subplot(312)
plt.imshow(sat,cmap=plt.cm.gray)
plt.subplot(313)
plt.imshow(val,cmap=plt.cm.gray)
```

```python
color = ('b','g','r')
plt.clf()
for i,col in enumerate(color):
    histr = cv2.calcHist([imageHSV],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
histHSV = cv2.calcHist([imageHSV], [0, 1],None, [180,256], [0, 180, 0, 256])
```

```python
plt.clf()
plt.imshow(histHSV,interpolation = 'nearest')
plt.show()
```
