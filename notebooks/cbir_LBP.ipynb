```python
import cv2
from skimage import feature
import numpy as np
import matplotlib.pyplot as plt
import imutils
from imutils import paths
```

```python
image1 = cv2.imread('Objects/faces/image_0326.jpg')
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
plt.imshow(image1)
gray1 = cv2.cvtColor(image1, cv2.COLOR_RGB2GRAY)
```

```python
numPoints=8
radius=1
lbp = feature.local_binary_pattern(gray1, numPoints,radius, method="uniform")
n_bins = int(lbp.max() + 1)
hist, _ = np.histogram(lbp, density=True, bins=n_bins, range=(0, n_bins))
#(hist, _) = np.histogram(lbp.ravel(),bins=np.arange(0, numPoints + 3),range=(0, numPoints + 2))
#hist = hist.astype("float")
#hist /= (hist.sum() + 1e-7)
```

```python
plt.imshow(lbp,cmap='gray')
```

```python
#plt.hist(hist,density=True, bins=26)
plt.hist(lbp.ravel(), density=True, bins=n_bins, range=(0, n_bins),facecolor='0.5');
```

```python
def rotateImage(image, angle):
  image_center = tuple(np.array(image.shape[1::-1]) / 2)
  rot_mat = cv2.getRotationMatrix2D(image_center, angle, 1.0)
  result = cv2.warpAffine(image, rot_mat, image.shape[1::-1], flags=cv2.INTER_LINEAR)
  return result
```

```python
rotated=rotateImage(image1,180)
```

```python
plt.imshow(rotated)
```

```python
gray1r = cv2.cvtColor(rotated, cv2.COLOR_RGB2GRAY)
lbpr = feature.local_binary_pattern(gray1r, numPoints,radius, method="uniform")
n_bins = int(lbpr.max() + 1)
histr, _ = np.histogram(lbpr, density=True, bins=n_bins, range=(0, n_bins))
```
